```python
import pandas as pd
import numpy as np
```

```python
TS_similarity= pd.read_csv("TS_Similarity_223.csv")
TS_similarity
```

```python
test_data_heldouttest= pd.read_csv("Heldout_test_predictions_earlydetection.csv")
test_data_heldouttest
```

```python
test_data_heldouttest = pd.merge(test_data_heldouttest, TS_similarity)
test_data_heldouttest
```

```python
test_data_heldouttest_5=test_data_heldouttest[test_data_heldouttest["TOXICITY"]==1].reset_index(drop=True)
test_data_heldouttest_5
```

```python
k1= test_data_heldouttest_5[test_data_heldouttest_5["mfp_earlydetections"] == True][["smiles_r",  "ts_nearest"]]
k1["Model"] = "Using Structural Features"
k1
```

```python
len(k1)
```

```python
k2= test_data_heldouttest_5[test_data_heldouttest_5["allfeatures_earlydetections"] == True][["smiles_r",  "ts_nearest"]]
k2["Model"] = "DILIPredictor"
k2
```

```python
len(k2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from statannotations.Annotator import Annotator
```

```python
k = pd.concat([k1, k2]).reset_index(drop=True)
k
```

```python
k.groupby("Model").describe()
```

```python
fig, ax = plt.subplots(figsize=(10, 5), dpi=300)
sns.kdeplot(x='ts_nearest', data=k, hue='Model', common_norm=False)
plt.title("Kernel Density Function")
```

```python
plt.figure(dpi= 200)
sns.displot(k, x="ts_nearest", hue="Model", element="step", height = 5)
```
